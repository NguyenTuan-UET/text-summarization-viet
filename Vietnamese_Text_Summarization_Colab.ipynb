## ğŸš€ 6. Cháº¡y tÃ³m táº¯t vÄƒn báº£n

Thá»±c hiá»‡n tÃ³m táº¯t vÄƒn báº£n vÃ  trÃ­ch xuáº¥t tá»« khÃ³a tá»« cÃ¡c vÄƒn báº£n máº«u.

```python
def analyze_text(text, title):
    """HÃ m phÃ¢n tÃ­ch vÃ  hiá»ƒn thá»‹ káº¿t quáº£ tÃ³m táº¯t vÄƒn báº£n"""
    print("="*70)
    print(f"ğŸ“Š PHÃ‚N TÃCH: {title}")
    print("="*70)
    
    # Hiá»ƒn thá»‹ vÄƒn báº£n gá»‘c
    print("\nğŸ“„ VÄ‚N Báº¢N Gá»C:")
    print("-" * 50)
    print(text.strip())
    
    # TÃ³m táº¯t vÄƒn báº£n (3 cÃ¢u quan trá»ng nháº¥t)
    print("\nğŸ“ TÃ“M Táº®T (3 cÃ¢u quan trá»ng nháº¥t):")
    print("-" * 50)
    summary = textrank.summarize_text_compound(text)
    for i, sentence in enumerate(summary, 1):
        print(f"{i}. {sentence.strip()}")
    
    # TrÃ­ch xuáº¥t tá»« khÃ³a
    print("\nğŸ” Tá»ª KHÃ“A QUAN TRá»ŒNG (Top 10):")
    print("-" * 50)
    keywords = textrank.get_only_keywords(text)
    sorted_keywords = sorted(keywords.items(), key=lambda x: x[1], reverse=True)
    
    for i, (keyword, score) in enumerate(sorted_keywords[:10], 1):
        print(f"{i:2d}. {keyword:<15} (Ä‘iá»ƒm: {score:.4f})")
    
    print("\n" + "="*70 + "\n")
    return summary, keywords

# PhÃ¢n tÃ­ch vÄƒn báº£n Ä‘áº§u tiÃªn
summary1, keywords1 = analyze_text(text_vietnam, "Giá»›i thiá»‡u vá» Viá»‡t Nam")
```

```python
# PhÃ¢n tÃ­ch vÄƒn báº£n thá»© hai
summary2, keywords2 = analyze_text(text_ai, "Vá» cÃ´ng nghá»‡ AI")
```

```python
# PhÃ¢n tÃ­ch vÄƒn báº£n thá»© ba
summary3, keywords3 = analyze_text(text_education, "Vá» giÃ¡o dá»¥c")
```

```python
# Tá»•ng há»£p káº¿t quáº£
print("ğŸ“ˆ Tá»”NG Há»¢P Káº¾T QUáº¢ PHÃ‚N TÃCH")
print("="*80)

results = [
    ("Viá»‡t Nam", summary1, keywords1),
    ("CÃ´ng nghá»‡ AI", summary2, keywords2), 
    ("GiÃ¡o dá»¥c", summary3, keywords3)
]

for topic, summary, keywords in results:
    print(f"\nğŸ·ï¸  CHá»¦ Äá»€: {topic}")
    print(f"ğŸ“ Sá»‘ cÃ¢u tÃ³m táº¯t: {len(summary)}")
    print(f"ğŸ” Sá»‘ tá»« khÃ³a: {len(keywords)}")
    
    # Top 3 tá»« khÃ³a quan trá»ng nháº¥t
    top_keywords = sorted(keywords.items(), key=lambda x: x[1], reverse=True)[:3]
    top_words = [word for word, score in top_keywords]
    print(f"ğŸ¯ Top 3 tá»« khÃ³a: {', '.join(top_words)}")
    print("-" * 50)
```
